```python
import pandas as pd

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head()
```

```python
train = train.dropna()
```

```python
def convert_label(label):

    if label <= 2:
        return 0   # Fake
    else:
        return 1   # Real

train["label"] = train["Labels"].apply(convert_label)
```

```python
train["label"].value_counts()
```

```python
train["content"] = train["Text"] + " " + train["Text_Tag"]
```

```python
X = train["content"]
y = train["label"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")

X_vector = vectorizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_vector,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2000)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```

```python
def predict_news(text, topic):

    input_data = text + " " + topic

    vector = vectorizer.transform([input_data])

    prediction = model.predict(vector)

    if prediction[0] == 0:
        return "Fake News"
    else:
        return "Real News"
```

```python
predict_news(
    "The government built a secret alien base",
    "politics"
)
```
